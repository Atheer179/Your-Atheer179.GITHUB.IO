```python
import numpy as np

def particle_filter(num_particles, measurements, motion_noise=0.08, measure_noise=0.20):
    particles = np.random.uniform(0, 10, num_particles)
    weights   = np.ones(num_particles) / num_particles
    estimates = []

    for z in measurements:
        particles += np.random.normal(0, motion_noise, num_particles)
        weights    = np.exp(-0.5 * ((particles - z) / measure_noise) ** 2)
        weights   /= weights.sum()
        estimates.append(np.sum(weights * particles))
        indices    = np.random.choice(num_particles, size=num_particles, p=weights)
        particles  = particles[indices]

    return estimates

# Test it
true_states  = [5 + 0.1*i for i in range(20)]
measurements = [x + np.random.normal(0, 0.3) for x in true_states]
estimates    = particle_filter(200, measurements)

for t, e in zip(true_states, estimates):
    print(f"True: {t:.2f}  Estimate: {e:.2f}  Error: {abs(t-e):.2f}")
```
